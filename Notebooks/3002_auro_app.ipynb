### Homografa
La homografa es una transformacin que asigna los puntos de un punto al punto correspondiente en otra imagen. La homografa es una matriz de 33:

![](./fig/2024-06-26-19-02-32.png)

Si 2 puntos no estn en el mismo plano entonces tenemos que usar 2 homgrafos. De manera similar, para *n* planos, tenemos que usar *n* homgrafos. Si tenemos ms homgrafos, entonces debemos manejarlos todos correctamente. Por eso utilizamos la coincidencia de funciones.

**Coincidencia de caractersticas:** la coincidencia de caractersticas significa encontrar caractersticas correspondientes de dos conjuntos de datos similares en funcin de una distancia de bsqueda.

**Homografa:** Para detectar la homografa del objeto tenemos que obtener la matriz y usar la funcin ``findHomography()`` para obtener la homografa del objeto. 

 La computadora no puede encontrar su homgrafo si no tiene suficientes puntos de correspondencia. Por lo tanto, necesitamos encontrar suficientes puntos de correspondencia para encontrar la homografa del objeto. Cuando intentamos cambiar o mover el objeto en otra direccin, los humanos pueden ver objetos cercanos ms grandes que objetos lejanos, aqu la perspectiva est cambiando. Esto se llama **transformacin de perspectiva**.

```python
import cv2
import cv2.aruco as aruco
import numpy as np
import os
def findArucoMarkers(img, markerSize = 6, totalMarkers=250, draw=True):
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    key = getattr(aruco, f'DICT_{markerSize}X{markerSize}_{totalMarkers}')
    arucoDict = aruco.getPredefinedDictionary(key)
    arucoParam = aruco.DetectorParameters()
    detector = cv2.aruco.ArucoDetector(arucoDict, arucoParam)
    bboxs, ids, rejected = detector.detectMarkers(gray)
    # print(ids)
    if draw:
        aruco.drawDetectedMarkers(img, bboxs)
    return [bboxs, ids]

def arucoAug(bbox, id, img, imgAug, drawId = True):
    tl = bbox[0][0][0], bbox[0][0][1]
    tr = bbox[0][1][0], bbox[0][1][1]
    br = bbox[0][2][0], bbox[0][2][1]
    bl = bbox[0][3][0], bbox[0][3][1]
    h, w, c = imgAug.shape
    pts1 = np.array([tl, tr, br, bl])
    pts2 = np.float32([[0,0], [w,0], [w,h], [0,h]])
    matrix, _ = cv2.findHomography(pts2, pts1)
    imgout = cv2.warpPerspective(imgAug, matrix, (img.shape[1], img.shape[0]))
    return imgout
cap = cv2.VideoCapture(0)
imgAug = cv2.imread("mini.png")

while True:
    success, img = cap.read()
    arucofound = findArucoMarkers(img)
     # loop through all the markers and augment each one
    if  len(arucofound[0])!=0:
        for bbox, id in zip(arucofound[0], arucofound[1]):
            img = arucoAug(bbox, id, img, imgAug)
    cv2.imshow('img',img)
    k = cv2.waitKey(30) & 0xff
    if k == 27:
        break
cap.release()
cv2.destroyAllWindows()
```

Un polgono convexo es un polgono en el que los segmentos de lnea entre los puntos no van hacia adentro y los vrtices del polgono apuntan hacia afuera. Los ngulos interiores miden menos de 180. OpenCV proporciona el mtodo ``cv2.fillConvexPoly()`` para dibujar un polgono convexo relleno en una imagen. 

```python
import cv2
import cv2.aruco as aruco
import numpy as np
import os
def findArucoMarkers(img, markerSize = 6, totalMarkers=250, draw=True):
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    key = getattr(aruco, f'DICT_{markerSize}X{markerSize}_{totalMarkers}')
    arucoDict = aruco.getPredefinedDictionary(key)
    arucoParam = aruco.DetectorParameters()
    detector = cv2.aruco.ArucoDetector(arucoDict, arucoParam)
    bboxs, ids, rejected = detector.detectMarkers(gray)
    # print(ids)
    if draw:
        aruco.drawDetectedMarkers(img, bboxs)
    return [bboxs, ids]

def arucoAug(bbox, id, img, imgAug, drawId = True):
    tl = bbox[0][0][0], bbox[0][0][1]
    tr = bbox[0][1][0], bbox[0][1][1]
    br = bbox[0][2][0], bbox[0][2][1]
    bl = bbox[0][3][0], bbox[0][3][1]
    h, w, c = imgAug.shape
    pts1 = np.array([tl, tr, br, bl])
    pts2 = np.float32([[0,0], [w,0], [w,h], [0,h]])
    matrix, _ = cv2.findHomography(pts2, pts1)
    imgout = cv2.warpPerspective(imgAug, matrix, (img.shape[1], img.shape[0]))
    cv2.fillConvexPoly(img, pts1.astype(int), (0, 0, 0))
    imgout = img + imgout
    return imgout
cap = cv2.VideoCapture(0)
imgAug = cv2.imread("mini.png")
while True:
    success, img = cap.read()
    arucofound = findArucoMarkers(img)
    # loop through all the markers and augment each one
    if len(arucofound[0])!=0:
        for bbox, id in zip(arucofound[0], arucofound[1]):
            img = arucoAug(bbox, id, img, imgAug)
    cv2.imshow('img',img)
    k = cv2.waitKey(30) & 0xff
    if k == 27:
        break
cap.release()
cv2.destroyAllWindows()
```
